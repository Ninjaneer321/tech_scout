```python
nsf_awards_json_api = 'http://api.nsf.gov/services/v1/awards.json'
keywords = 'keyword=data analytics'

# Make the URL request
res = requests.get(nsf_awards_json_api + '?' + keywords)

# Get the awards data
json_response = res.json()["response"]
print(f'type(awards): {type(json_response)}')
awards = json_response["award"]
```

```python
nsf_df = pd.json_normalize(awards)
nsf_df
```

```python
nsf_df['Date2'] = nsf_df['date'].apply(lambda x: datetime.datetime.strptime(x, '%m/%d/%Y').strftime('%Y-%m-%d'))
nsf_df
```

```python
nsf_df['NAME'] = nsf_df['piFirstName'] + " " + nsf_df['piLastName']
nsf_df
```

```python
DA_nsfdf = pd.DataFrame()
DA_nsfdf["Name"] = nsf_df["NAME"]
DA_nsfdf["Organization"] = nsf_df["awardeeName"]
DA_nsfdf["Project"] = nsf_df["title"]
DA_nsfdf["Funding"] = nsf_df["fundsObligatedAmt"]
DA_nsfdf
```

```python
NIH_NSF = DA_nsfdf.append(NIH_df)
NIH_NSF
```

```python
NIH_NSF['Funding']=NIH_NSF['Funding'].astype(float)
NIH_NSF.dtypes
```

```python
agg_name = NIH_NSF[['Name','Organization','Funding']].groupby(['Organization','Name']).agg('sum')
agg_name_v1 = agg_name.reset_index()
agg_name_v1.columns = ["Organization","Name","Funding"]
agg_name_v1
```
