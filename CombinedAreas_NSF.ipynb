```python
# Test of simple scraper and entity extractor.
import requests
import datetime
# import packages
import pandas as pd
import numpy as np
import requests
from pandas.io.json import json_normalize #special package in pandas
import json
from matplotlib.pyplot import *
%matplotlib inline
```

```python
nsf_awards_json_api = 'http://api.nsf.gov/services/v1/awards.json'
keywords=['keyword=ai','keyword=cybersecurity','keyword=quantum','keyword=data analytics','keyword=modeling and simulation']
```

```python
combined_areas = pd.DataFrame()

for keyword in keywords:
    res = requests.get(nsf_awards_json_api + '?' + keyword)
    json_response = res.json()["response"]
    awards = json_response["award"]
    rdf = pd.json_normalize(awards)
    rdf["Area"] = keyword
    rdf["Topic"] = rdf["Area"].str[8:]
    combined_areas = combined_areas.append(rdf)
    

    
```

```python
combined_areas
```

```python
df = combined_areas.copy(deep=True)
df
```

```python
df = df.drop(['Area'], axis=1)
df
```
