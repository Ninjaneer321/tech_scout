```python
api_key = ""
endpoint = "https://api.nal.usda.gov/pubag/rest/search/?query=abstract:quantum"
params = {"api_key": api_key}
response = requests.get(endpoint, params=params)
res = response.json()
```

```python
df = pd.json_normalize(res["response"],record_path=['docs'])
df
```

```python
keywords=['ai','cybersecurity','quantum','data analytics','modeling and simulation']
```

```python
combined_areas = pd.DataFrame()

for keyword in keywords:
    a_params = {"api_key": api_key, "per_page":100}
    endpoint = "https://api.nal.usda.gov/pubag/rest/search/?query=abstract:quantum"
    response = requests.get(endpoint, params=a_params)
    res = response.json()
    rdf = pd.json_normalize(res["response"],record_path=['docs'])
    rdf["Area"] = keyword
    combined_areas = combined_areas.append(rdf)
```
